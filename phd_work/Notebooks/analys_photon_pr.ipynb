```python
# Размер батча (настройте под вашу память)
batch_size = 100  
n_components = 3

# 1. Инициализация масштабировщика и инкрементального PCA
scaler = StandardScaler()
ipca = IncrementalPCA(n_components=n_components)

# 2. Первый проход: обучение масштабировщика на всех данных
print("Обучение масштабировщика...")
for batch in tqdm(np.array_split(latent_list, len(latent_list) // batch_size)):
    scaler.partial_fit(batch)  # Инкрементальный расчет mean и std

# 3. Второй проход: применение масштабирования и обучение PCA
print("Обучение PCA...")
for batch in tqdm(np.array_split(latent_list, len(latent_list) // batch_size)):
    batch_scaled = scaler.transform(batch)  # Применяем масштабирование к батчу
    ipca.partial_fit(batch_scaled)  # Обучение PCA на масштабированных данных

# 4. Преобразование всех данных
print("Финализация...")
latent_list_scaled = scaler.transform(latent_list)  # Масштабируем весь датасет
transformed_data = ipca.transform(latent_list_scaled)  # Применяем PCA
```
